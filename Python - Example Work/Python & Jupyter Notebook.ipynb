#### Question 1d <span style="color:red">(10 marks)</span>
Build and train a linear regression model. Print its weights and accuracy. Which is the most important feature and how to interpret its contribution to your prediction?

<b>Q1d answer</b>: 

```python
import numpy as np # Import numpy as np
from sklearn.model_selection import train_test_split # Import train test split
from sklearn.linear_model import LinearRegression # Import linear regression
from sklearn.model_selection import cross_val_score # Import cross val score
from sklearn.metrics import mean_squared_error # Import mean squared error
from sklearn import metrics # Import metrics

################################################################################################################################

x = movies[["Year", "Gross_income", "Budget", "Screens", "Sequel_no", "Sentiment", "Views", "Likes",
            "Dislikes", "Comments", "Actors_Followers"]] # Select Features
y = movies["Ratings"] # Select Target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0) # Split into training and test

model = LinearRegression() # Create a Linear Regression model

model.fit(x_train, y_train) # Train the model

predictions = model.predict(x_test) # Make predictions on the test set

################################################################################################################################

print("\033[1m" + "Weights (Coefficients):" + "\033[0m")
print(model.coef_) # Print the model's weights (coefficients)

print("\033[1m" + "\nIntercept:" + "\033[0m")
print(model.intercept_) # Print the model's intercept

cv_score = cross_val_score(model, x, y, cv=5, scoring="neg_mean_squared_error").mean()*-1 # Calculate cross val score
print("\033[1m" + "\nCross-validation scores:" + "\033[0m")
print(cv_score) # Print cross val score

mse = mean_squared_error(y_test, predictions) # Calculate mean squared error
print("\033[1m" + "\nMean Squared Error:" + "\033[0m")
print(mse) # Print mean squared error

print("\n-------------------------------------------------------------------------------------------------------------------\n")

################################################################################################################################

x = movies[["Budget"]] # Select Features

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0) # Split into training and test

model = LinearRegression() # Create a Linear Regression model
model.fit(x_train, y_train) # Train the model
predictions = model.predict(x_test) # Make predictions on the test set

plt.scatter(x_test["Budget"], y_test, color="blue", label="Actual Ratings") # Add the actual ratings to the scatter graph
plt.scatter(x_test["Budget"], predictions, color="red", label="Predicted Ratings") # Add the predictions to the scatter graph

x_values = np.linspace(x_test["Budget"].min(), x_test["Budget"].max(), 100).reshape(-1, 1) # Generate a sequence of x-values

y_values = model.predict(x_values) # Calculate the corresponding y-values using the model's coefficients

plt.plot(x_values, y_values, color="green", label="Regression Line") # Plot the regression line
plt.xlabel("Budget") # Label x axis
plt.ylabel("Ratings") # Label y axis
plt.title("Ratings Prediction VS Actual Ratings") # Title the graph
plt.legend() # Add the key to the graph
plt.show() # Show the graph

################################################################################################################################

x = movies[["Gross_income"]] # Select Features

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0) # Split into training and test

model = LinearRegression() # Create a Linear Regression model
model.fit(x_train, y_train) # Train the model
predictions = model.predict(x_test) # Make predictions on the test set

plt.scatter(x_test["Gross_income"], y_test, color="blue", label="Actual Ratings") # Add the actual ratings to the scatter graph
plt.scatter(x_test["Gross_income"], predictions, color="red", label="Predicted Ratings") # Add predictions to the scatter graph

x_values = np.linspace(x_test["Gross_income"].min(), x_test["Gross_income"].max(), 100).reshape(-1, 1) # Sequence of x-values

y_values = model.predict(x_values) # Calculate the corresponding y-values using the model's coefficients

plt.plot(x_values, y_values, color="green", label="Regression Line") # Plot the regression line
plt.xlabel("Gross Income") # Label x axis
plt.ylabel("Ratings") # Label y axis
plt.title("Ratings Prediction VS Actual Ratings") # Title the graph
plt.legend() # Add a key to the graph
plt.show() # Show the graph

################################################################################################################################
```
